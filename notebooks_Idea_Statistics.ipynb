# Sampling
**This study estimates a population proportion using a simple random sample (SRS) of size 75 with binary outcomes (injury = yes/no). That maps directly to a binomial sampling algorithm.**


```python
#sampling
#Curency amounts	- Mean
#Heights / weights - Mean
#Yes / No outcomes	Proportion

#bernolli variable - categorical numerical variable used in sampling 
import numpy as np

# Example: 75 sampled outcomes
sample = np.random.choice([0, 1], size=75)

# estimate proportion of head injuries
sample_proportion = sample.mean()

sample_proportion
```

```python
#This study estimates a population proportion using a simple random sample (SRS) of size 75 with binary outcomes (injury = yes/no). That maps directly to a binomial sampling algorithm.
# Algorithm (conceptual version)

# Goal: estimate

# 𝑝
# =
# true proportion of head injuries
# p=true proportion of head injuries

# Procedure:

# Step 1
# Define population injury probability 𝑝
# p (example assumption: maybe 0.40)

# Step 2
# Take random sample of 75 cars

# Step 3
# Record injury outcome for each dummy:

# 1 = head injury
# 0 = no head injury

# Step 4
# Compute sample proportion

# 𝑝^= number of injuries 75
# p^= number of injuries /75

import numpy as np

# assumed true probability of injury (unknown in real life)
p = 0.40

# sample size
n = 75

# simulate sample outcomes
sample = np.random.binomial(1, p, size=n)

# compute sample proportion
sample_proportion = sample.mean()

print(sample)
print("Estimated proportion:", sample_proportion)
	
```

```python
#Full statistical sampling simulation algorithm of varaibility (produces a different outcome every time it is repeated)
import numpy as np
import matplotlib.pyplot as plt

# assumed true probability of head injury
p = 0.40

# number of cars sampled per study
n = 75

# number of repeated simulated studies
num_trials = 1000

estimates = []

for _ in range(num_trials):
    sample = np.random.binomial(1, p, size=n)
    estimates.append(sample.mean())

print("Average estimated proportion:", np.mean(estimates))

# visualize sampling distribution
plt.hist(estimates, bins=20)
plt.title("Sampling distribution of estimated injury proportion")
plt.xlabel("Estimated proportion")
plt.ylabel("Frequency")
plt.show()
```

```python
# How to make results repeatable (important for research & debugging)
# If you want the same result every time, set a random seed:

import numpy as np
import matplotlib.pyplot as plt

# pseudo-random number generator (PRNG)
# This is an algorithm that produces numbers that look random but are actually deterministic.
# Meaning:
# same starting point → same sequence every time
# That starting point is the seed.
np.random.seed(42)   # fixes randomness

p = 0.40
n = 75
num_trials = 1000

estimates = []

for _ in range(num_trials):
    sample = np.random.binomial(1, p, size=n)
    estimates.append(sample.mean())

print("Average estimated proportion:", np.mean(estimates))

# Now the output will be identical each run.

# This is called reproducibility 🔬
```
